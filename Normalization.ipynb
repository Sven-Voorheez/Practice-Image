```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gamma_correction(I_in, c, gamma):
    # Step 1: Normalize image
    I_norm = I_in.astype('float32') / 255.0
    
    # Step 2: Apply power law
    I_gamma = c * (I_norm ** gamma)
    
    # Step 3: Scale back to 0â€“255
    I_out = np.clip(I_gamma * 255, 0, 255).astype('uint8')
    
    return I_out
```

```python
# Read image
image = cv2.imread("/Users/210719/Downloads/argues.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply gamma correction
output = gamma_correction(image, c=1, gamma=0.5)

# Show result
plt.imshow(output)
plt.title("Gamma Corrected")
plt.axis("off")
plt.show()
```

```python
def contrast_stretching(I_in, s_min, s_max):
    
    # Convert to float
    I = I_in.astype('float32')
    
    # Find r_min and r_max
    r_min = np.min(I)
    r_max = np.max(I)
    
    # Apply contrast stretching
    I_out = ((I - r_min) / (r_max - r_min)) * (s_max - s_min) + s_min
    
    return I_out
```

```python
image = cv2.imread("/Users/210719/Downloads/argues.jpg", 0)

output = contrast_stretching(image, 0, 255)
output = output.astype('uint8')

plt.imshow(output, cmap='gray')
plt.title("Contrast Stretched")
plt.axis("off")
plt.show()
```

```python
image = cv2.imread("/Users/210719/Downloads/argues.jpg", 0)

# Normalize
image_norm = image.astype('float32') / 255

output = contrast_stretching(image_norm, 0, 1)

plt.imshow(output, cmap='gray')
plt.title("Contrast Stretched (Normalized)")
plt.axis("off")
plt.show()
```
