```python
import cv2
import matplotlib.pyplot as plt

# Read image in grayscale
image = cv2.imread('argues.jpg', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

plt.figure(figsize=(10,5))

# Original Image
plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Original Histogram
plt.subplot(2,2,2)
plt.hist(image.ravel(), 256, [0,256])
plt.title('Histogram (Original)')

# Equalized Image
plt.subplot(2,2,3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

# Equalized Histogram
plt.subplot(2,2,4)
plt.hist(equalized_image.ravel(), 256, [0,256])
plt.title('Histogram (Equalized)')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read image
color_image = cv2.imread('argues.jpg')

# Split channels
(b, g, r) = cv2.split(color_image)

# Equalize each channel
equalized_b = cv2.equalizeHist(b)
equalized_g = cv2.equalizeHist(g)
equalized_r = cv2.equalizeHist(r)

# Merge channels back
equalized_color_image = cv2.merge([equalized_b, equalized_g, equalized_r])

plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
plt.title('Original Color Image')
plt.axis('off')

# Equalized Image
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(equalized_color_image, cv2.COLOR_BGR2RGB))
plt.title('Equalized Color Image (Channels Separately)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from skimage import exposure
import matplotlib.pyplot as plt

# Read two grayscale images
image_source = cv2.imread('argues.jpg', cv2.IMREAD_GRAYSCALE)
image_target = cv2.imread('Melissa.webp', cv2.IMREAD_GRAYSCALE)

# Perform histogram matching
matched_image = exposure.match_histograms(image_source, image_target)

# Display images
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(image_source, cmap='gray')
plt.title('Source Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(image_target, cmap='gray')
plt.title('Target Image')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(matched_image, cmap='gray')
plt.title('Matched Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from skimage import exposure
import matplotlib.pyplot as plt

# Read source and target images
image_source = cv2.imread('', cv2.IMREAD_GRAYSCALE)
image_target = cv2.imread('', cv2.IMREAD_GRAYSCALE)

# Step 1: Equalize the source image
equalized_source = cv2.equalizeHist(image_source)

# Step 2: Perform histogram matching on the equalized image
matched_after_equalization = exposure.match_histograms(equalized_source, image_target)

# Display results
plt.figure(figsize=(15,5))

plt.subplot(2,3,1)
plt.imshow(image_source, cmap='gray')
plt.title('Original Source')
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(equalized_source, cmap='gray')
plt.title('Equalized Source')
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(matched_after_equalization, cmap='gray')
plt.title('Equalized + Matched')
plt.axis('off')

plt.subplot(2,3,4)
plt.hist(image_source.ravel(), 256, [0,256])
plt.title('Histogram (Original)')

plt.subplot(2,3,5)
plt.hist(equalized_source.ravel(), 256, [0,256])
plt.title('Histogram (Equalized)')

plt.subplot(2,3,6)
plt.hist(matched_after_equalization.ravel(), 256, [0,256])
plt.title('Histogram (Equalized + Matched)')

plt.show()
```
